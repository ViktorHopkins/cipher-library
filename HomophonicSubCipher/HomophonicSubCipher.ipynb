```python
import random 

class CharacterKeeper:
    # Initialize a data structure to keep track of clear text characters and their associated encrypted characters
    def __init__(self):
        self.clearChar = ''
        self.cryptChar = []
        self.cryptCharLeft = []

class HomophonicSubCipher:
    # Initialize the clear text alphabet and the list for character mappings between clear text and encrypted characters
    clearAlphabet = list("abcdefghijklmnopqrstuvwxyzåäö !.,?")
    cryptList = []

    def fillCryptList():
        HomophonicSubCipher.cryptList.clear()
        rng = random.Random()

        # Create mappings for each character in the clear alphabet
        for char in HomophonicSubCipher.clearAlphabet:
            c = CharacterKeeper()
            c.clearChar = char
            HomophonicSubCipher.cryptList.append(c)

        # Assign random encrypted character indexes for each clear text character
        for _ in range(2500):
            num = rng.randint(0, len(HomophonicSubCipher.clearAlphabet) - 1)
            HomophonicSubCipher.cryptList[num].cryptChar.append(_)

        # Check if each clear text character has enough encrypted character mappings
        for c in HomophonicSubCipher.cryptList:
            if len(c.cryptChar) < 50:
                return False
        return True

    def encryptChar(c):
        # Encrypt a clear text character by mapping it to a randomly selected encrypted character index
        for i in range(len(HomophonicSubCipher.clearAlphabet)):
            if HomophonicSubCipher.clearAlphabet[i] == c:
                ck = HomophonicSubCipher.cryptList[i]

                if not ck.cryptCharLeft:
                    ck.cryptCharLeft = list(ck.cryptChar)

                rng = random.Random()
                num = rng.randint(0, len(ck.cryptCharLeft) - 1)
                returnNumber = ck.cryptCharLeft[num]
                ck.cryptCharLeft.pop(num) # Remove the selected encrypted character from the available list
                return returnNumber
        return -1

    def decryptChar(number):
        # Decrypt a encrypted character index by finding the corresponding clear text character
        for i in range(len(HomophonicSubCipher.cryptList)):
            if number in HomophonicSubCipher.cryptList[i].cryptChar:
                return HomophonicSubCipher.clearAlphabet[i]
        return ' '

    def encryptString(input_string):
        # Encrypts a clear text string by mapping each character to a list of encrypted character indexes
        encrypted = []
        for char in input_string:
            encryptedChar = HomophonicSubCipher.encryptChar(char)
            encrypted.append(encryptedChar)
        return encrypted

    def decryptString(input_list):
        # Decrypts a list of encrypted character indexes
        decrypted = []
        for number in input_list:
            decryptedChar = HomophonicSubCipher.decryptChar(number)
            decrypted.append(decryptedChar)
        return ''.join(decrypted)

    def getCryptList():
        return HomophonicSubCipher.cryptList
```

```python
# Create an instance of the HomophonicSubCipher class for filling the list for character mappings, encrypting and decrypting
HomophonicSubCipher.fillCryptList()
```
