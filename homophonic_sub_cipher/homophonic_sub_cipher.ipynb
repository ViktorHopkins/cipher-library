```python
import random 

class CharacterKeeper:
    # Initialize a data structure to keep track of clear text characters and their associated encrypted characters
    def __init__(self):
        self.clear_char = ''
        self.crypt_char = []
        self.crypt_char_left = []

class HomophonicSubCipher:
    # Initialize the clear text alphabet and the list for character mappings between clear text and encrypted characters
    clear_alphabet = list("abcdefghijklmnopqrstuvwxyzåäö !.,?")
    clear_list = []

    def fill_clear_list():
        HomophonicSubCipher.clear_list.clear()
        rng = random.Random()

        # Create mappings for each character in the clear alphabet
        for char in HomophonicSubCipher.clear_alphabet:
            c = CharacterKeeper()
            c.clear_char = char
            HomophonicSubCipher.clear_list.append(c)

        # Assign random encrypted character indexes for each clear text character
        for _ in range(2500):
            num = rng.randint(0, len(HomophonicSubCipher.clear_alphabet) - 1)
            HomophonicSubCipher.clear_list[num].crypt_char.append(_)

        # Check if each clear text character has enough encrypted character mappings
        for c in HomophonicSubCipher.clear_list:
            if len(c.crypt_char) < 50:
                return False
        return True

    def encrypt_char(c):
        # Encrypt a clear text character by mapping it to a randomly selected encrypted character index
        for i in range(len(HomophonicSubCipher.clear_alphabet)):
            if HomophonicSubCipher.clear_alphabet[i] == c:
                ck = HomophonicSubCipher.clear_list[i]

                if not ck.crypt_char_left:
                    ck.crypt_char_left = list(ck.crypt_char)

                rng = random.Random()
                num = rng.randint(0, len(ck.crypt_char_left) - 1)
                returnNumber = ck.crypt_char_left[num]
                ck.crypt_char_left.pop(num) # Remove the selected encrypted character from the available list
                return returnNumber
        return -1

    def decrypt_char(number):
        # Decrypt a encrypted character index by finding the corresponding clear text character
        for i in range(len(HomophonicSubCipher.clear_list)):
            if number in HomophonicSubCipher.clear_list[i].crypt_char:
                return HomophonicSubCipher.clear_alphabet[i]
        return ' '

    def encrypt_string(input_string):
        # Encrypts a clear text string by mapping each character to a list of encrypted character indexes
        encrypted = []
        for char in input_string:
            encrypted_char = HomophonicSubCipher.encrypt_char(char)
            encrypted.append(encrypted_char)
        return encrypted

    def decrypt_string(input_list):
        # Decrypts a list of encrypted character indexes
        decrypted = []
        for number in input_list:
            decrypted_char = HomophonicSubCipher.decrypt_char(number)
            decrypted.append(decrypted_char)
        return ''.join(decrypted)

    def get_clear_list():
        return HomophonicSubCipher.clear_list
```

```python
# Create an instance of the HomophonicSubCipher class for filling the list for character mappings, encrypting and decrypting
HomophonicSubCipher.fill_clear_list()
```
