```python
class VigenèreCipher:
    def __init__(self):
        # Initialize the clear text alphabet
        self.clearAlphabet = list("abcdefghijklmnopqrstuvwxyz")

    def encrypt(self, keyword, text):
        keyword = list(keyword)
        message = list(text)

        keywordIndex = 0
        returnString = ""

        # Loop through message
        for i in range(len(message)): 
            if message[i] == ' ':
                returnString += " "
                continue

            charPosition = 0

            # Loop through clear alphabet and find the position of the keyword character, increment the charPosition
            for k in range(len(self.clearAlphabet)): 
                if keyword[keywordIndex] == self.clearAlphabet[k]:
                    charPosition = k
                    break

            clearPosition = 0

            # Check if message character matche character in the clear alphabet
            for j in range(len(self.clearAlphabet)): 
                if message[i] == self.clearAlphabet[j]:
                    clearPosition = j
                    break

            # Calculates the new position
            position = clearPosition + charPosition 

            # Out of range position gets wrapped around
            if position >= len(self.clearAlphabet): 
                position -= len(self.clearAlphabet)

            # Get the character in the new position and append the character to the result
            tmp = self.clearAlphabet[position] 
            returnString += tmp
            keywordIndex += 1

            # Create a "Keywordstream" where the keyword wraps around when it has reached its end
            if keywordIndex == len(keyword): 
                keywordIndex = 0

        return returnString

    def decrypt(self, keyword, text):
        # Convert the keyword and text into lists of characters
        keyword = list(keyword)
        message = list(text)

        keywordIndex = 0
        returnString = ""

         # Loop through each character in the message, appending space characters
        for i in range(len(message)):
            if message[i] == ' ':
                returnString += " "
                continue

            charPosition = 0

            for k in range(len(self.clearAlphabet)):
                if keyword[keywordIndex] == self.clearAlphabet[k]:
                    charPosition = k
                    break

            cryptPosition = 0

            # Loop through the clear alphabet to find the position of the message character, store position where message character matches matches clear alphabet
            for j in range(len(self.clearAlphabet)):
                if message[i] == self.clearAlphabet[j]:
                    cryptPosition = j
                    break

            # Calculate the new position for the character in the decrypted message
            position = cryptPosition - charPosition

            # Handle case when the position is negative by wrapping it around to the end of the alphabet
            if position < 0:
                position += len(self.clearAlphabet)

            # Get the character in the new position and append it to the result
            tmp = self.clearAlphabet[position]
            returnString += tmp
            keywordIndex += 1

             # Create a "Keywordstream" where the keyword wraps around when it has reached its end
            if keywordIndex == len(keyword):
                keywordIndex = 0

        return returnString
```
